```python
def opinion_extractor_1_1(aspect_token, parsed_sentence):
    opinions = list()
    
    def crazy_stuff(token):
        for advmod_child in token.children:
            if advmod_child.dep_ == 'advmod':
                if has_neg(token):
                    for neg_child in token.children:
                        if neg_child.dep_ == 'neg':
                            opinions.append("not-" + advmod_child.orth_ + "-" + token.orth_)
                else:
                    opinions.append(advmod_child.orth_ + "-" + token.orth_) 
    
    def neg(token):
        for neg_child in token.children:                        
            if neg_child.dep_ == 'neg':
                opinions.append("not-" + token.orth_)
    
    for token in parsed_sentence:
        if token.pos_ == 'NOUN' and token.orth_ == aspect_token:
            for child in token.children:
                if child.dep_ == 'amod':
                    if has_advmod(child):
                        for advmod_child in child.children:                        
                            if advmod_child.dep_ == 'advmod':
                                opinions.append(advmod_child.orth_ + "-" + child.orth_)
                    elif has_neg(token):
                        for neg_child in token.children:                        
                            if neg_child.dep_ == 'neg':
                                opinions.append("not-" + child.orth_)
                    else:
                        opinions.append(child.orth_)
            if token.dep_ == 'nsubj':
                for child in token.head.children:
                    if child.dep_ == 'acomp':
                        for conj_child in child.children:
                            if conj_child.dep_ == 'conj':
                                if has_advmod(conj_child):
                                    crazy_stuff(conj_child)
                                else:                                            
                                    opinions.append(conj_child.orth_)
                                    if list(conj_child.children):
                                        for conj_child_2 in conj_child.children:
                                            if conj_child_2.dep_ == 'conj':
                                                opinions.append(conj_child_2.orth_)
                        if has_advmod(child):
                            crazy_stuff(child)
                        elif has_neg(child): 
                            neg(child)
                        else: 
                            opinions.append(child.orth_)            
                            
    return opinions
    

def has_advmod(token):
    for child in token.children:
        if child.pos_ == 'ADV' and child.dep_ == 'advmod':
            return True

def has_neg(token):
    for child in token.children:
        if child.pos_ == 'ADV' and child.dep_ == 'neg':
            return True
        
aspect_words = ["plot","characters","cinematography","dialogue"]

for doc in core:
    for word in aspect_words:
        if word in doc[1]:
            print("Results for word '{}'".format(word))
            print("\tSentence: {}".format(doc[1]))
            print("\t", opinion_extractor(word, nlp(doc[1])))
            print("--------------------------------------------------------------------------------------------") 
            
"""
def opinion_extractor_1_2(aspect_token, parsed_sentence):
    opinions = list()
    
    def adj(child):
        if child.pos_ == 'ADJ':            
            if has_neg(child.head) or has_neg(child) and not has_advmod(child):
                opinions.append("not-" + child.orth_)
            elif has_advmod(child):            
                for kid in child.children:                
                    adv_mod(child, kid)
            else:
                opinions.append(child.orth_)
            for kid in child.subtree:
                if kid.dep_ == 'conj':
                    if has_advmod(kid):
                        for child_kid in kid.children:
                            adv_mod(kid, child_kid)
                    else:
                        opinions.append(kid.orth_)

    def has_conj(child):
        for kid in child.children:
            if kid.dep_ == 'conj':
                return True
    
    def adv_mod(child_1, child_2):
        if child_2.dep_ == 'advmod' or child_2.dep_ == 'conj' and child_1.dep_ == 'advmod':
            if has_neg(child_1):
                opinions.append("not-" + child_2.orth_ + "-" + child_1.orth_)
            else:
                opinions.append(child_2.orth_ + "-" + child_1.orth_)
                
    def has_advmod(token):
        for child in token.children:
            if child.dep_ == 'advmod':
                return True

    def has_neg(token):
        for child in token.children:
            if child.dep_ == 'neg':
                return True
    
    for token in parsed_sentence:
        if token.pos_ == 'NOUN' and token.orth_ == aspect_token:
            for child in token.children:
                adj(child)
            if token.dep_ == 'nsubj':
                for child in token.head.children:
                    adj(child)
                    
    return opinions
    
aspect_words = ["plot","characters","cinematography","dialogue"]

for doc in core:
    for word in aspect_words:
        if word in doc[1]:
            print("Results for word '{}'".format(word))
            print("\tSentence: {}".format(doc[1]))
            print("\t", opinion_extractor(word, nlp(doc[1])))
            print("--------------------------------------------------------------------------------------------")     
"""            
```
