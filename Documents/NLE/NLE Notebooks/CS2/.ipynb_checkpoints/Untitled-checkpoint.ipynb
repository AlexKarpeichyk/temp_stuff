```python
# retrieving a gender map from file
def create_gender_map(dict_reader):
    names_info = defaultdict(lambda: {"gender":"", "freq": 0.0})
    for row in input_file:
        name = row["name"].lower()
        if names_info[name]["freq"] < float(row["freq"]):
            names_info[name]["gender"] = row["gender"] 
            names_info[name]["freq"] = float(row["freq"])
    gender_map = defaultdict(lambda: "unknown")
    for name in names_info:
        gender_map[name] = names_info[name]["gender"]
    return gender_map

# extending gender map for titles
def extend_gender_map(gender_map):
    # initialyze a dictionary for male and female versions of various titles
    titles = {"male": ["mr", "master", "uncle", "sir", "prince", "archduke", "grand duke", 
                       "duke", "marquis", "count", "baron", "emperor", "king", "pope", 
                       "brother", "father", "cardinal", "captain", "agent"], 
              "female": ["mrs", "ms", "miss", "maid", "madam", "aunt", "princess", 
                         "archduchess", "grand duchess", "duchess", "marchioness", "marquise", 
                         "countess", "baroness", "lady", "queen", "empress", "sister"]}    
    for gender in titles.keys(): # iterate over dictionary of titles
        for count in range(len(titles[gender])):
            gender_map[titles[gender][count]] = gender # extend given gender map with title and its gender
    return gender_map # return the updated gender map

# identifying gender of a character given a name and a gender map
def guess_gender(name, gender_map): 
    tokenised_name = name.split() # split name into tokens (title, name, surname, etc.)
    if gender_map[name] == "unknown": # check if gender of a name is defined in gender map
        if len(tokenised_name) > 1: # check if name is longer than one word
            first_name = tokenised_name[0] # store first token of the name (first name or title) in a variable
            last_name = tokenised_name[len(tokenised_name) - 1] # store the last token (most likely surname)
            gender_map[last_name] = gender_map[first_name] # create a new entry in gender map with last name as key and gender of first name as value
            return gender_map[first_name] # return gender of first name
    return gender_map[name] # return gender of name 

# getting a list of names and their counts in a document (novel)
def named_entity_counts(document,named_entity_label):
    occurrences = [ent.string.strip() for ent in document.ents
                   if ent.label_ == named_entity_label and ent.string.strip()]
    return Counter(occurrences)
```

Functions create_gender_map and named_entity_counts were given to us in Topic 7 and were not edited by me for the coursework, so I will focus on the functions that I introduced.

**Function extend_gender_map**

This function is simply extends the given gender map (created by the create_gender_map function) to allow for a more accurate performance of the gender classifier, i.e. to be able to define gender of a character when he/she is referred using a title (Mr, Mrs, etc). I introduced a dictionary that maps a gender to a tuple, that contains gender variations of different titles, or titles that are unique (specific) to that gender. Using a for loop I add a title value (from a tuple) and map it to its gender key and finally return the updated map.

**Function guess_gender**

This function returns gender of a given name argument if it's defined in given gender map. If the gender is unknown, I check for the length of the name to be longer than 1 (when first and last name is used, or when a title is used). I then split the name into tokens to access first name/title and surname separately. I assign first name/title and last name to variables and create a new entry in the gender map with surname as key, and gender of the first name as value. This is done to deal with situations when the text reveals the gender of character first (Mr. Smith, John Smith) and only the surname is used (Smith). The way I did it also allows to switch gender of a surname if the text contains two characters with the same surname but different genders. Thus, I assume that when a gender is revealed, the next mentioning of that character (using surname only) referres to the character whose gender has been revealed previously. Finally, the function returns gender of the first name (when character is refered using more than just name). 

The rest of the code, presented below just runs my gender classifier on a set of data and output the results.

```python
input_file = csv.DictReader(open("names.csv"))
gender_map = create_gender_map(input_file)
texts = {parsed_emma: "Emma", parsed_moby: "Moby Dick", parsed_alice: "Alice", parsed_sense: "Sense", parsed_persuasion: "Persuasion"}
entity_type = 'PERSON'
number_of_entities = 30
for text in texts.keys():
    names_with_gender = [(name, guess_gender(name.lower(), extend_gender_map(gender_map))) 
                         for name, count in named_entity_counts(text,entity_type).most_common(number_of_entities)]
    cnt = 0
    for name, gender in names_with_gender:
        if not gender == "unknown":
            cnt += 1
    print("Results for", texts[text])
    print("Genders defined:", cnt*100/number_of_entities, "%")
    display(pd.DataFrame(names_with_gender,columns=["Name","Gender"]))
    print("-----------------------------------------")
```

## Section 3


Apart from one way of characterising how a person is portrayed by a novelist that was offered to us in topic 8, I was able to introduce a few more ways of doing that. The given idea involved retrieving verbs that are in relation with the character entities of a novel, i.e. what is done to them and what they do. One of alternative ideas is to also derive adjectives that describe a character in one way or another. Thus, the adjectives would either be connected to a noun directly, or via copula. E.g. 'John is smart/tall/etc...', 'This is smart John'...
Another feature that I added is also include an adverbial modifier if a relevant adjective has one (or many). Here's a good example:



Poor unknown someone...

And one more feature I introduced is when a character is described using a noun (attribute) or a noun with an adjectival/adverbial modifier. E.g. 'John is a genius..', 'John is a complete idiot..'.

```python
# Code Block 1

# retrieve entities from a novel
def get_entities_in(parsed_novel,entity_type):
    return [ent.text.strip().lower() for ent in parsed_novel.ents 
            if ent.label_ == entity_type and ent.text.strip()]

# get a list of a given number of main characters in a given novel
def get_main_characters(parsed_novel, num_characters):
    all_characters = get_entities_in(parsed_novel,"PERSON")
    return [person for person,count in Counter(all_characters).most_common(num_characters)]

# get collection of features for characters in a set of novels
def get_interesting_contexts_characters(novels, rels, num_characters, num_common_verbs):
    
    # checking for characters having links to verbs (what they do, or what is done to them)
    def of_interest_verb(ent, rels, main_characters):
        return (ent.text.strip().lower() in main_characters 
                and ent.label_ == 'PERSON' 
                and ent.root.head.pos_ == 'VERB'
                and ent.root.dep_ in rels)
    
    # checking for characters having links to adjectives via verb (copula) (John is smart)
    def of_interest_copula_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and ent.root.head.pos_ == 'VERB'
                and adj.pos_ == 'ADJ'
                and adj in ent.root.head.children)
    
    # checking for characters to have direct links to adjectives (smart John)
    def of_interest_direct_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and adj.pos_ == 'ADJ'
                and adj in ent.root.children)
    
    # checking for characters to have links to nouns that describe them (John is an idiot)
    def of_interest_copula_noun(ent, noun, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and ent.root.head.pos_ == 'VERB'
                and noun.pos_ == 'NOUN'
                and noun in ent.root.head.children)
    
    # checking for a token to have negative children
    def has_negative(token):
        for child in token.children:
            if child.dep_ == 'neg':
                return "not "
        return ""
    
    # checking for a token to have adverbial modifiers
    def has_advmod(token):
        adv_mod = ""
        for child in token.subtree:
            if child.dep_ == 'advmod' and child.pos_ == 'ADV':
                adv_mod += child.orth_.lower() + " "
        return adv_mod
    
    # checking for a token to have adjectival modifiers
    def has_amod(token):
        a_mod = ""
        for child in token.children:
            if child.dep_ == 'amod' and child.pos_ == 'ADJ' or child.pos_ == 'ADV':
                a_mod += child.orth_.lower() + " "
        return a_mod
    
    # Code Block 2
    
    contexts = defaultdict(Counter)
    for parsed_novel in novels:
        # list of all verbs in a novel
        verbs = [verb.string.strip().lower() for verb in parsed_novel if verb.pos_ == 'VERB' and verb.string.strip()]
        # most common verbs in a novel
        most_common_verbs = [verb[0] for verb in Counter(verbs).most_common(num_common_verbs)]
        # get a list of a specified number of main characters
        main_characters = get_main_characters(parsed_novel, num_characters)
        for ent in parsed_novel.ents: # going over entities (characters) in a novel
            # check if value returned by a method is true + check that a verb is not one of the most common verbs
            if of_interest_verb(ent, rels, main_characters) and not ent.root.head.lemma_ in most_common_verbs:
                # add character name to the contexts table with a feature and make it negative if needed (has_negative function)
                contexts[ent.text.strip().lower()][has_negative(ent.root.head) + ent.root.head.lemma_] += 1
            # go over children of an entity's head (verb/copula)
            for adj in ent.root.head.children:
                # check if value returned by a method is true + if relationship between entity's head and its child is 'acomp'
                if of_interest_copula_adj(ent, adj, main_characters) and adj.dep_ == 'acomp':
                    # add character name to the contexts table with a feature, make it negative if needed (has_negative function), add adverbial modifier if needed (has_advmod function)
                    contexts[ent.text.strip().lower()][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            # go over children of an entity (character)
            for adj in ent.root.children:
                # check if value returned by a method is true + if relationship between entity and its child is 'amod'
                if of_interest_direct_adj(ent, adj, main_characters) and adj.dep_ == 'amod':
                    # add character name to the contexts table with a feature, make it negative if needed (has_negative function), add adverbial modifier if needed (has_advmod function)
                    contexts[ent.text.strip().lower()][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            # go over children of an entity (character)
            for noun in ent.root.children:
                # check if value returned by a method is true + if relationship between entity and its child is 'attr'
                if of_interest_copula_noun(ent, noun, main_characters) and noun.dep_ == 'attr':
                    # add character name to the contexts table with a feature, make it negative if needed (has_negative function), add adjectival modifier if needed (has_amod function)
                    contexts[ent.text.strip().lower()][has_negative(adj) + has_amod(noun) + noun.lemma_] += 1
    return contexts

novels_1 = {parsed_Pemberton_1, parsed_Pemberton_2, parsed_Pemberton_3, parsed_Pemberton_4, parsed_Pemberton_5}
number_of_characters_per_text = 4
num_common_verbs = 20
target_rels = {'nsubj', 'dobj'}
target_contexts = get_interesting_contexts_characters(novels_1, target_rels, number_of_characters_per_text, num_common_verbs)
display(pd.DataFrame.from_dict(target_contexts).applymap(lambda x: '' if math.isnan(x) else x))
```

**Code Block 1**

I introduced a few more methods (of_interest...), that are just versions of the of_interest method that was given to us initially, and they simply check for a certain feature (adjectival modifier, attribute) to be linked with one of the main characters, which insures relevant features will be present in the contexts table. Additional functions (has_negative, etc.) are used to add a 'complete' feature, when, for example, an adjective has an adverbial modifier and a negative child (so instead of just 'smart', the feature in the output table will look 'not so smart').

**Code Block 2**

Building up the table of features (contexts dictionary) is just a matter of going over characters (entities) of each novel in a set and check if character has some feature by calling one of the 'of_interest..' methods and checking for relevant relations between entity and feature. When adding character name and feature to the table we also update the feature string (contexts[..]['feature here'] += 1) by calling additional methods ('has' methods), which return a negative, adverbial or adjectival modifier as a string, which is added to the feature (adjective or attribute) itself and will appear in the output table.

The code below is an edited version of the code above, for showing how characters are portrayed when referred using a pronoun ("he", "she", "his" and "her").

```python
# Code Block 1

# get collection of features for pronouns in a set of novels
def get_interesting_contexts_pronouns(novels, rels, num_characters, num_common_verbs):
    
    # checking for pronouns having links to verbs (what they do, or what is done to them)
    def of_interest(ent, rels, main_characters):
        return (ent.text.strip().lower() in main_characters 
                and ent.root.pos_ == 'PRON' 
                and ent.root.head.pos_ == 'VERB'
                and ent.root.dep_ in rels)
    
    # checking for pronouns having links to adjectives via verb (copula) (John is smart)
    def of_interest_copula_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.root.pos_ == 'PRON' 
                and ent.root.head.pos_ == 'VERB'
                and adj.pos_ == 'ADJ'
                and adj in ent.root.head.children)
    
    # checking for pronouns to have direct links to adjectives (smart John)
    def of_interest_direct_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.root.pos_ == 'PRON' 
                and adj.pos_ == 'ADJ'
                and adj in ent.root.children)
    
    # checking for pronouns to have links to nouns that describe them (John is an idiot)
    def of_interest_copula_noun(ent, noun, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.root.pos_ == 'PRON' 
                and ent.root.head.pos_ == 'VERB'
                and noun.pos_ == 'NOUN'
                and noun in ent.root.head.children)
    
    # checking for a token to have negative children
    def has_negative(token):
        for child in token.children:
            if child.dep_ == 'neg':
                return "not "
        return ""
    
    # checking for a token to have adverbial modifiers
    def has_advmod(token):
        adv_mod = ""
        for child in token.subtree:
            if child.dep_ == 'advmod' and child.pos_ == 'ADV':
                adv_mod += child.orth_.lower() + " "
        return adv_mod
    
    # checking for a token to have adjectival modifiers
    def has_amod(token):
        a_mod = ""
        for child in token.children:
            if child.dep_ == 'amod' and child.pos_ == 'ADJ' or child.pos_ == 'ADV':
                a_mod += child.orth_.lower() + " "
        return a_mod
    
    # Code Block 2
    
    contexts = defaultdict(Counter)
    for parsed_novel in novels:
        # get list of all verbs in a novel
        verbs = [verb.string.strip().lower() for verb in parsed_novel if verb.pos_ == 'VERB' and verb.string.strip()]
        # get most common verbs in a novel
        most_common_verbs = [verb[0] for verb in Counter(verbs).most_common(num_common_verbs)]
        # use pronouns instead of character names
        main_characters = ["he", "she", "her", "his"]
        # going over noun chunks of a novel
        for ent in parsed_novel.noun_chunks:
            if of_interest(ent, rels, main_characters) and not ent.root.head.lemma_ in most_common_verbs:
                contexts[ent.root.text.strip().lower()][has_negative(ent.root.head) + ent.root.head.lemma_] += 1
            for adj in ent.root.head.children:
                if of_interest_copula_adj(ent, adj, main_characters) and adj.dep_ == 'acomp':
                    contexts[ent.root.text.strip().lower()][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            for adj in ent.root.children:
                if of_interest_direct_adj(ent, adj, main_characters) and adj.dep_ == 'amod':
                    contexts[ent.root.text.strip().lower()][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            for noun in ent.root.children:
                if of_interest_copula_noun(ent, noun, main_characters) and noun.dep_ == 'attr':
                    contexts[ent.root.text.strip().lower()][has_negative(adj) + has_amod(noun) + noun.lemma_] += 1
    return contexts

novels_1 = {parsed_Pemberton_1, parsed_Pemberton_2, parsed_Pemberton_3, parsed_Pemberton_4, parsed_Pemberton_5}
num_common_verbs = 20
target_rels = {'nsubj', 'dobj'}
target_contexts = get_interesting_contexts_pronouns(novels_1, target_rels, number_of_characters_per_text, num_common_verbs)
display(pd.DataFrame.from_dict(target_contexts).applymap(lambda x: '' if math.isnan(x) else x))
```

As for Pemberton novels, we may try to alter the cosine similarity by making sure that the there's no imbalance (i.e. use the same number of male and female characters). For that purpose I introduced function get_main_characters_even.

```python
# retrieve entities from a novel
def get_entities_in(parsed_novel,entity_type):
    return [ent.text.strip().lower() for ent in parsed_novel.ents 
            if ent.label_ == entity_type and ent.text.strip()]

# get a list of a given number of main characters in a given novel (with even number of male and female characters)
def get_main_characters_even(parsed_novel, num_characters):
    all_characters = get_entities_in(parsed_novel,"PERSON")
    main_characters = list()
    main_characters_male = []
    main_characters_female = []
    for person, count in Counter(all_characters).most_common(num_characters):
        if guess_gender(person, extend_gender_map(gender_map)) == 'male':
            main_characters_male.append(person)
    for person, count in Counter(all_characters).most_common(num_characters):
        if guess_gender(person, extend_gender_map(gender_map)) == 'female':
            main_characters_female.append(person)
    if len(main_characters_female) < len(main_characters_male):
        if len(main_characters_female) == 1:
            main_characters_male = main_characters_male[len(main_characters_male)-1:]
        else:    
            for count in range(len(main_characters_male) - len(main_characters_female)):
                del main_characters_male[count]
    if len(main_characters_female) > len(main_characters_male):
        for count in range(len(main_characters_female) - len(main_characters_male)):
            del main_characters_female[count]
    
    for m in main_characters_male:
        main_characters.append(m)
        
    for f in main_characters_female:
        main_characters.append(f)
        
    print('Number of male characters: ', len(main_characters_male))
    print('Number of female characters: ', len(main_characters_female))
    
    return main_characters

def get_interesting_contexts_gender_even(novels, rels, num_characters, num_common_verbs):
    
    def of_interest(ent, rels, main_characters):
        return (ent.text.strip().lower() in main_characters 
                and ent.label_ == 'PERSON' 
                and ent.root.head.pos_ == 'VERB'
                and ent.root.dep_ in rels)
    
    def of_interest_copula_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and ent.root.head.pos_ == 'VERB'
                and adj.pos_ == 'ADJ'
                and adj in ent.root.head.children)
    
    def of_interest_direct_adj(ent, adj, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and adj.pos_ == 'ADJ'
                and adj in ent.root.children)
    
    def of_interest_copula_noun(ent, noun, main_characters):
        return (ent.text.strip().lower() in main_characters
                and ent.label_ == 'PERSON'
                and ent.root.head.pos_ == 'VERB'
                and noun.pos_ == 'NOUN'
                and noun in ent.root.head.children)
    
    def has_negative(token):
        for child in token.children:
            if child.dep_ == 'neg':
                return "not "
        return ""
    
    def has_advmod(token):
        adv_mod = ""
        for child in token.subtree:
            if child.dep_ == 'advmod' and child.pos_ == 'ADV':
                adv_mod += child.orth_.lower() + " "
        return adv_mod
    
    def has_amod(token):
        a_mod = ""
        for child in token.children:
            if child.dep_ == 'amod' and child.pos_ == 'ADJ' or child.pos_ == 'ADV':
                a_mod += child.orth_.lower() + " "
        return a_mod
    
    contexts = defaultdict(Counter)
    for parsed_novel in novels:
        # get list of all verbs in a novel
        verbs = [verb.string.strip().lower() for verb in parsed_novel if verb.pos_ == 'VERB' and verb.string.strip()]
        # get most common verbs in a novel
        most_common_verbs = [verb[0] for verb in Counter(verbs).most_common(num_common_verbs)]
        main_characters = get_main_characters_even(parsed_novel, num_characters)
        for ent in parsed_novel.ents:
            if of_interest(ent, rels, main_characters) and not ent.root.head.lemma_ in most_common_verbs and not guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map)) == "unknown":
                contexts[guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map))][has_negative(ent.root.head) + ent.root.head.lemma_] += 1
            for adj in ent.root.head.children:
                if of_interest_copula_adj(ent, adj, main_characters) and adj.dep_ == 'acomp' and not guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map)) == "unknown":
                    contexts[guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map))][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            for adj in ent.root.children:
                if of_interest_direct_adj(ent, adj, main_characters) and adj.dep_ == 'amod' and not guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map)) == "unknown":
                    contexts[guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map))][has_negative(adj) + has_advmod(adj) + adj.lemma_] += 1
            for noun in ent.root.children:
                if of_interest_copula_noun(ent, noun, main_characters) and noun.dep_ == 'attr' and not guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map)) == "unknown":
                    contexts[guess_gender(ent.text.strip().lower(), extend_gender_map(gender_map))][has_negative(adj) + has_amod(noun) + noun.lemma_] += 1
    
    return contexts

novels_1 = {parsed_Pemberton_1, parsed_Pemberton_2, parsed_Pemberton_3, parsed_Pemberton_4, parsed_Pemberton_5}
all_novels = [novels_1]
number_of_characters_per_text = 20
num_common_verbs = 15
target_rels = {'nsubj', 'dobj'}
for novel_set in all_novels:   
    target_contexts = get_interesting_contexts_gender_even(novel_set, target_rels, number_of_characters_per_text, num_common_verbs)
    display(pd.DataFrame.from_dict(target_contexts).applymap(lambda x: '' if math.isnan(x) else x))
```

Now when we made sure that the number of male and female characters is the same, we may calculate the cosine similarity again. 

```python
from sklearn.metrics.pairwise import cosine_similarity

novels_1 = {parsed_Pemberton_1, parsed_Pemberton_2, parsed_Pemberton_3, parsed_Pemberton_4, parsed_Pemberton_5}

target_contexts_Pemberton = get_interesting_contexts_gender_even(novels_1, target_rels, number_of_characters_per_text, num_common_verbs)

female_counter_Pemberton = Counter(target_contexts_Pemberton["female"])
male_counter_Pemberton = Counter(target_contexts_Pemberton["male"])

def counters_to_feature_lists(female_counter, male_counter):
    combined = female_counter + male_counter
    female_list = [female_counter[key] for key in combined]
    male_list = [male_counter[key] for key in combined]
    return female_list, male_list

female_Pemberton, male_Pemberton = counters_to_feature_lists(female_counter_Pemberton, male_counter_Pemberton)

print('Pemberton novels: ', cosine_similarity([female_Pemberton], [male_Pemberton])[0,0])
```

And evidently, getting rid of imbalance in the number of characters for Pemberton novels reveals that the cosine similarity is actually a little bit smaller, than with the imbalance (original similarity: 0.675382181573), so the difference in the way Pemberton portrays his male and female characters is even more significant. 

This tells us that the imbalance can indeed affect the value of cosine similarity.
