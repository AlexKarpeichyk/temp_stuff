### Named Entity Recognition
Noun phrases in a document can refer to entities such as people, places, organisations, times, etc. These noun phrases are **entity mentions**.

Named Entity Recognition (NER) refers a process in which two things are achieved: 
1. identifying the location of a phrase within the text that is an entity mention, and 
2. determining the type of entity that the entity mention is referring to.

spaCy labels base noun phrases with their entity type. 

Here is a summary of spaCy's entity types.

TYPE | DESCRIPTION
:----: |---
`PERSON` | People, including fictional
`NORP` | Nationalities or religious or political groups
`FACILITY` | Buildings, airports, highways, bridges, etc
`ORG` | Companies, agencies, institutions, etc
`GPE` | Countries, cities, states
`LOC` | Non-GPE locations, mountain ranges, bodies of water
`PRODUCT` | Objects, vehicles, foods, etc (Not services)
`EVENT` | Named hurricanes, battles, wars, sports events, etc
`WORK_OF_ART` | Titles of books, songs, etc
`LAW` | Named documents made into laws
`LANGUAGE` | Any named language
`DATE` | Absolute or relative dates or periods.
`TIME` | Times smaller than a day
`PERCENT` | Percentage, including "%".
`MONEY` | Monetary values, including unit
`QUANTITY` | Measurements, as of weight or distance
`ORDINAL` | "first", "second", etc
`CARDINAL` | Numerals that do not fall under another type


More details can be found [here](https://spacy.io/docs/usage/entity-recognition#entity-types).

````python
rcParams['figure.figsize'] = 16, 4

def locations(target_token, document,num_bins): 
    list_of_indices = [ent.root.i for ent in document.ents if ent.string.strip().lower() == target_token]
    return pd.Series(np.histogram(list_of_indices, bins=num_bins)[0])

text = parsed_moby
number_of_bins = 100
characters = {'Starbuck', 'Stubb'}
pd.DataFrame(
    {name: locations(name.lower(),text,number_of_bins) 
     for name in characters}).plot(subplots=True)
````

## Exploring gender differences

We are now going to look at whether there are differences in the verbal contexts of characters based on gender. Our first task is to define a function that guesses the gender of a character based on their name.  

We have provided a csv file called `names.csv` that contains several thousand male and female names with information about how frequent each name is. 

We can read in the contents of this file with this line of python:  

```
input_file = csv.DictReader(open("names.csv"))
```

See [here](https://docs.python.org/3/library/csv.html) for description of `DictReader`.

Some of the names listed in `names.csv` appear as both the female and male, so the first thing we do is to establish which gender is more likely in these cases. To do this, we will use the function, `most_likely_gender`, defined as follows:

```
def most_likely_gender(dict_reader):
    names_info = defaultdict(lambda: defaultdict(str)) 
    for row in dict_reader:
        if names_info[row["name"].lower()]["freq"] < row["freq"]: # is this gender more frequent?
            names_info[row["name"].lower()]["gender"] = row["gender"] 
            names_info[row["name"].lower()]["freq"] = row["freq"]
    return names_info
    
```
`most_likely_gender` returns `names_info` which is a dictionary where the keys are all of the lowercase strings produced from the names that appear in `names.csv`. Each name maps to a dictionary that has just two keys: `"freq"` and `"gender"`. The key `"freq"` maps to the frequency of the name, and the key `"gender"` maps to the gender `"male"` or `"female"`.


The following line warrants an explanation.
```

```

````python
def most_likely_gender(dict_reader):
    names_info = defaultdict(lambda: defaultdict(str)) 
    for row in dict_reader:
        name = row["name"].lower()
        if names_info[name]["freq"] < row["freq"]: # is this gender more frequent?
            names_info[name]["gender"] = row["gender"] 
            names_info[name]["freq"] = row["freq"]            
    return names_info
    
def get_gender_map(file):
    input_file = csv.DictReader(open(file))
    names_details = most_likely_gender(input_file) 
    genders = defaultdict(str)
    for name in names_details:
        genders[name] = names_info[name]["gender"]
    return genders

gender_map = get_gender_map("names.csv")
other_male_names = ["captain","colonel","mr","sir"]
other_female_names = ["mrs","miss","ms"]
for name in other_male_names:
    gender_map[name] = "male"
for name in other_female_names:
    gender_map[name] = "female"

def guess_gender(name):
    if not gender_map[name]:
        name = name.split()[0]
    return gender_map[name]

guess_gender("jack")
````

````python
# %load solutions/

def get_most_common_verbs(texts,num_verbs_to_ignore):
    
    def get_all_verbs(documents):
        all_verbs = set()
        for document in documents:
            for token in document:
                if token.pos_ == 'VERB':
                    all_verbs.add((token.lemma_,token.prob))
        return list(all_verbs)

    def get_most_frequent(words,k):
        sorted_words = sorted(words,key=itemgetter(1),reverse=True)
        return sorted_words[:k]
    
    return [verb for verb,prob in get_most_frequent(get_all_verbs(texts),num_verbs_to_ignore)]

def get_interesting_contexts(documents,rels,num_people,too_common_verbs):
    
    def get_entities_in(document,entity_type):
        return [ent.string.strip().lower() for ent in document.ents 
                if ent.label_ == entity_type and ent.string.strip()]
    def get_main_characters(document,num_people):
        all_people = get_entities_in(document,"PERSON")
        return [person for person,count in Counter(all_people).most_common(num_people)]
    def of_interest(ent,rels,main_characters):
        return (ent.string.strip().lower() in main_characters 
                and ent.label_ == 'PERSON' 
                and ent.root.head.pos_ == 'VERB'
                and not ent.root.head.lemma_ in too_common_verbs
                and ent.root.dep_ in rels)  

    contexts = defaultdict(Counter)    
    for document in documents:
        main_characters = get_main_characters(document,num_people)
        for ent in document.ents:
            if of_interest(ent,rels,main_characters):
                contexts[ent.string.strip().lower()][ent.root.head.lemma_] += 1
    return contexts

def remove_non_target_contexts(target,non_target): 
    for person in non_target.keys():
        for verb in non_target[person]:
            del target[person][verb]
    return target

def split_contexts_by_gender(contexts):
    male_contexts = defaultdict(Counter)
    female_contexts = defaultdict(Counter)
    for person in contexts.keys():
        if guess_gender(person) == "male":
            male_contexts[person] = contexts[person]
        elif guess_gender(person) == "female":
            female_contexts[person] = contexts[person]
        else: # gender unknown
            pass
    return male_contexts,female_contexts

def aggregate_context_counts(contexts):
    aggregated_contexts = Counter()
    for person in contexts.keys():
        for verb in contexts[person]:
            aggregated_contexts[verb] += contexts[person][verb]
    return aggregated_contexts

def context_differences(contextsA,contextsB):
    contextsTemp = contextsA # we're going to change contextsA so let's make a copy
    contextsA.subtract(contextsB)
    contextsB.subtract(contextsTemp)
    return contextsA,contextsB

# configuration options
texts = {parsed_emma,parsed_persuasion,parsed_sense}
rels = {'nsubj','dobj'}
target_rels = {'nsubj'}
non_target_rels = rels - target_rels
number_of_people_per_text = 50
num_verbs_to_ignore = 20
num_verbs_to_show = 20

too_common_verbs = get_most_common_verbs(texts,num_verbs_to_ignore)
target_contexts = get_interesting_contexts(texts,target_rels,number_of_people_per_text,too_common_verbs)
non_target_contexts = get_interesting_contexts(texts,non_target_rels,number_of_people_per_text,too_common_verbs)
# only keep target contexts that aren't also non_target contexts
target_contexts = remove_non_target_contexts(target_contexts,non_target_contexts)
# divide up contexts by gender
male_contexts,female_contexts = split_contexts_by_gender(target_contexts)
# aggregate all contexts for male and female characters
aggregated_male_contexts = aggregate_context_counts(male_contexts)
aggregated_female_contexts = aggregate_context_counts(female_contexts)
# find gender differences
male_deltas,female_deltas = gender_differences(aggregated_male_contexts,aggregated_female_contexts)
male_verbs = [verb for verb,count in male_deltas.most_common(num_verbs_to_show)]
female_verbs = [verb for verb,count in female_deltas.most_common(num_verbs_to_show)]
display(pd.DataFrame(list(zip_longest(male_verbs,female_verbs)),columns=["Male","Female"]))
````

````python
# %load solutions/
from operator import itemgetter, attrgetter, methodcaller

def get_all_verbs(documents):
    all_verbs = set()
    for document in documents:
        for token in document:
            if token.pos_ == 'VERB':
                all_verbs.add((token.lemma_,token.prob))
    return list(all_verbs)

def get_most_frequent(words,k):
    sorted_words = sorted(words,key=itemgetter(1),reverse=True)
    return sorted_words[:k]

def get_most_common_verbs(texts,num_verbs_to_ignore):
    return [verb for verb,prob in get_most_frequent(get_all_verbs(texts),num_verbs_to_ignore)]

def get_entities_in(document,entity_type):
    return [ent.string.strip().lower() for ent in document.ents 
            if ent.label_ == entity_type and ent.string.strip()]

def get_main_characters(document,num_people):
    all_people = get_entities_in(document,"PERSON")
    return [person for person,count in Counter(all_people).most_common(num_people)]

def wanted_context(ent,rels,main_characters):
    return (ent.string.strip().lower() in main_characters 
            and ent.label_ == 'PERSON' 
            and ent.root.head.pos_ == 'VERB'
            and not ent.root.head.lemma_ in common_verbs
            and ent.root.dep_ in rels)

def get_contexts(documents,rels,num_people,common_verbs):
    contexts = defaultdict(Counter)
    for document in documents:
        main_characters = get_main_characters(document,num_people)
        for ent in document.ents:
            if wanted_context(ent,rels,main_characters):
                contexts[ent.string.strip().lower()][ent.root.head.lemma_] += 1
    return contexts

def remove_non_target_contexts(target,non_target): 
    for person in non_target.keys():
        for verb in non_target[person]:
            del target[person][verb]
    return target

def split_contexts_by_gender(contexts):
    male_contexts = defaultdict(Counter)
    female_contexts = defaultdict(Counter)
    for person in contexts.keys():
        if guess_gender(person) == "male":
            male_contexts[person] = contexts[person]
        elif guess_gender(person) == "female":
            female_contexts[person] = contexts[person]
        else: # gender unknown
            pass
    return male_contexts,female_contexts

def aggregate_context_counts(contexts):
    aggregated_contexts = Counter()
    for person in contexts.keys():
        for verb in contexts[person]:
            aggregated_contexts[verb] += contexts[person][verb]
    return aggregated_contexts

def context_differences(contextsA,contextsB):
    contextsTemp = contextsA # we're going to change contextsA so let's make a copy
    contextsA.subtract(contextsB)
    contextsB.subtract(contextsTemp)
    return contextsA,contextsB

# configuration options
texts = {parsed_emma,parsed_persuasion,parsed_sense}
rels = {'nsubj','dobj'}
target_rels = {'nsubj'}
non_target_rels = rels - target_rels
number_of_people = 50
num_verbs_to_ignore = 0
num_verbs_to_show = 20

common_verbs = get_most_common_verbs(texts,num_verbs_to_ignore)
target_contexts = get_contexts(texts,target_rels,number_of_people,common_verbs)
non_target_contexts = get_contexts(texts,non_target_rels,number_of_people,common_verbs)
# only keep target contexts that aren't also non_target contexts
target_contexts = remove_non_target_contexts(target_contexts,non_target_contexts)
# divide up contexts by gender
#male_contexts,female_contexts = split_contexts_by_gender(dobj_contexts)
male_contexts,female_contexts = split_contexts_by_gender(target_contexts)
# aggregate all contexts for male characters
collapsed_male_contexts = aggregate_context_counts(male_contexts)
# aggregate all contexts for female characters
collapsed_female_contexts = aggregate_context_counts(female_contexts)
# find gender differences
male_deltas,female_deltas = gender_differences(collapsed_male_contexts,collapsed_female_contexts)
male_verbs = [verb for verb,count in male_deltas.most_common(num_verbs_to_show)]
female_verbs = [verb for verb,count in female_deltas.most_common(num_verbs_to_show)]
display(pd.DataFrame(list(zip_longest(male_verbs,female_verbs)),columns=["Male","Female"]))
````

Find adjectives used to describe characters in a book

````python
def verbsForCharacters(text, character):
    sents = [sent for sent in text.sents if character in sent.string]
    charWords = []
    for sent in sents: 
        for word in sent: 
            if character in word.string: 
                charWords.append(word)
    charAdjectives = []
    for word in charWords: 
        # Start walking up the list of ancestors 
        # Until we get to the first verb. 
        for ancestor in word.ancestors: 
            if ancestor.pos_.startswith('V'): 
                charAdjectives.append(ancestor.lemma_.strip())
    return Counter(charAdjectives).most_common(20)

emma_words = verbsForCharacters(parsed_emma,"Emma")
elton_words = verbsForCharacters(parsed_emma,"Elton")
harriet_words = verbsForCharacters(parsed_emma,"Harriet")
verbsToMatrix(emma_words)
````

````python
def verbsToMatrix(verbCounts): 
    """ 
    Takes verb counts given by verbsForCharacters 
    and makes Pandas Series out of them, suitabe for combination in 
    a DataFrame. 
    """
    return pd.Series({t[0]: t[1] for t in verbCounts})

verbsDF = pd.DataFrame({'Emma': verbsToMatrix(emma_words), 
                        'Elton': verbsToMatrix(elton_words), 
                        'Harriet': verbsToMatrix(harriet_words)}).fillna(0)
verbsDF.plot(kind='bar', figsize=(14,4))
````
